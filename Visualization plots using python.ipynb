# Raincloud plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips.head()
```

```python
df = [tips['total_bill'].values, tips['tip'].values]
```

```python
f, ax = plt.subplots(figsize=(7, 5))
# Boxplot
bp = ax.boxplot(df, patch_artist = True, vert = False)

# violin plot
violin_colors = ['blue', 'red']
vp = ax.violinplot(df, points=500,showmeans=False, showextrema=False, showmedians=False, vert=False)

for idx, b in enumerate(vp['bodies']):
    # Show only top half of the plot
    b.get_paths()[0].vertices[:, 1] = np.clip(b.get_paths()[0].vertices[:, 1], idx+1, idx+2)
    # Change the color
    b.set_color(violin_colors[idx])
    
# Scatterplot
scatter_colors = ['tomato', 'darksalmon']
for idx, features in enumerate(df):
    # Add jitter effect so the features do not overlap on the y-axis
    y = np.full(len(features), idx + .8)
    idxs = np.arange(len(y))
    out = y.astype(float)
    out.flat[idxs] += np.random.uniform(low=-.05, high=.05, size=len(idxs))
    y = out
    plt.scatter(features, y, s=.3, c=scatter_colors[idx])
    
plt.yticks(np.arange(1,3,1), ['Total bill amount', 'Tips amount'])  # Set text labels.
plt.xlabel('Values')
plt.title("Raincloud plot")
plt.show()
```
