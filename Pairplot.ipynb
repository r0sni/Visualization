```python
#Import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Read the data and store it in dataframe
data_file = 'iris.data'
col_names = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Class']
data = pd.read_csv(data_file, names = col_names)
data.head()
```

```python
data['Class'].value_counts()
```

```python
# Set the style and context for the plot
sns.set_style("white")
sns.set_context("notebook", font_scale=1.2)

# Create a pair plot with a custom color palette
g = sns.pairplot(data, hue='Class', diag_kind='kde', palette='Set2', plot_kws={'s': 50, 'edgecolor': 'w'}, 
                 markers=["o", "s", "D"])

# Adjust the space between the plots
g.fig.subplots_adjust(top=0.95, right=0.95)

# Add a title to the figure
g.fig.suptitle('Pair Plot with Customizations', fontsize=18, weight='bold', y=1.02)

# Adjust the padding between the axes labels and the plots
for ax in g.axes.flatten():
    ax.set_xlabel(ax.get_xlabel(), labelpad=15)
    ax.set_ylabel(ax.get_ylabel(), labelpad=15)
    ax.grid(True, linestyle='--', color='gray', alpha=0.5)
    ax.set_facecolor('#f7f7f7')
        
#Remove default legend   
g._legend.remove()

# Add a customized legend to the grid
g.add_legend(title='Species')
sns.move_legend(g, frameon=True, facecolor='lightgrey', edgecolor='black',
                loc='center right', bbox_to_anchor=(1, 0.5))
    
# Show the plot
plt.show()
```
